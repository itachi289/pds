```python
import pandas as pd

# Creating the raw data from the table in the assignment
data = {
    'Height': [65.8, 71.5, 69.4, 68.2, 67.8, 68.7, 69.8, 70.1, 67.9, 66.8],
    'Weight': [112, 136, 153, 142, 144, 123, 141, 136, 112, 120],
    'Age': [30, 19, 45, 22, 29, 50, 51, 23, 17, 39],
    'Grip Strength': [30, 31, 29, 28, 24, 26, 22, 20, 19, 31],
    'Frailty': ['N', 'N', 'N', 'Y', 'Y', 'N', 'Y', 'Y', 'N', 'N']
}

# Creating a DataFrame from the raw data
df = pd.DataFrame(data)

# Step 1: Check for null values and fill them
# For numeric columns, fill null values with the mean of the column
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())

# For the 'Frailty' column (categorical), fill null values with the mode (most frequent value)
df['Frailty'] = df['Frailty'].fillna(df['Frailty'].mode()[0])

# Step 2: Check for duplicate values and remove them (if any)
df.drop_duplicates(inplace=True)

# Step 3: Convert categorical 'Frailty' column from 'Y'/'N' to binary (1/0)
df['Frailty'] = df['Frailty'].map({'Y': 1, 'N': 0})

# Step 4: Save the processed DataFrame to CSV for further analysis
df.to_csv('frailty_data_processed.csv', index=False)

# Display the processed DataFrame
df.head()

```
