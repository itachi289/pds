Use this data to perform the following:

a. Look for the missing values in all the columns and either impute them (replace with mean,median, or mode) or drop them. Justify your action for this task.

```python
import pandas as pd
```

```python
#Load the dataset
from google.colab import files
uploaded = files.upload()
data = pd.read_csv('train.csv')
```

```python
# Check for missing values in all columns
missing_values_before = data.isnull().sum()

# Display the missing values count for each column
print("Missing Values Before Imputation:")
print(missing_values_before)
```

```python
# Impute numerical columns with the median (for columns with skewed distribution or outliers)
numerical_columns = data.select_dtypes(include=['float64', 'int64']).columns
for column in numerical_columns:
    data[column] = data[column].fillna(data[column].median())

# Impute categorical columns with the mode (most frequent value)
categorical_columns = data.select_dtypes(include=['object']).columns
for column in categorical_columns:
    data[column] = data[column].fillna(data[column].mode()[0])

# Verify if any missing values remain
missing_values_after = data.isnull().sum()

# Display the results
print("\nMissing Values After Imputation:")
print(missing_values_after)
```

B. Remove the units from some of the attributes and only keep the numerical values for example remove kmpl from “Mileage”, CC from “Engine”, bhp from “Power”, and lakh from

```python
# Remove 'kmpl' from 'Mileage' and convert to numeric
data['Mileage'] = data['Mileage'].replace(r' kmpl', '', regex=True)
data['Mileage'] = data['Mileage'].replace(r' km/kg', '', regex=True)  # in case of km/kg
data['Mileage'] = pd.to_numeric(data['Mileage'], errors='coerce')
```

```python
# Remove 'CC' from 'Engine' and convert to numeric (only the first part before 'CC')
data['Engine'] = data['Engine'].replace(r' CC', '', regex=True)
data['Engine'] = data['Engine'].replace(r'(\d+)', r'\1', regex=True)  # Keep numeric part only
data['Engine'] = pd.to_numeric(data['Engine'], errors='coerce')
```

```python
# Remove 'bhp' from 'Power' and convert to numeric
data['Power'] = data['Power'].replace(r' bhp', '', regex=True)
data['Power'] = pd.to_numeric(data['Power'], errors='coerce')
```

```python
# Remove 'Lakh' from 'New_Price' and convert to numeric (replace ' Lakh' with nothing)
data['New_Price'] = data['New_Price'].replace(r' Lakh', '', regex=True)
data['New_Price'] = data['New_Price'].replace(r',', '', regex=True)  # In case of commas in large numbers
data['New_Price'] = pd.to_numeric(data['New_Price'], errors='coerce')


# Display the cleaned data
data.head()
```

C)  Change the categorical variables (“Fuel_Type” and “Transmission”) into numerical one hot
encoded value.

```python

# One-hot encode 'Fuel_Type' and 'Transmission' columns
data=pd.get_dummies(data,columns=['Fuel_Type', 'Transmission'], drop_first=True)

# Display the first few rows of the updated data
data.head()
```

e) Perform select, filter, rename, mutate, arrange and summarize with group by operations (or
their equivalent operations in python) on this dataset.

```python
# Select specific columns
selected_data = data[['Name', 'Price', 'Mileage', 'Car_Age']]

# Display the selected data
selected_data.head()
```

```python
# Filter rows where car price is greater than 8
filtered_data = data[data['Price'] > 8]

# Display the filtered data
filtered_data.head()
```

```python
# Rename columns (equivalent to rename in R)
data.rename(columns={'Price':'Car_Price', 'Mileage':'Fuel_Efficiency'}, inplace=True)

# Display the renamed data
data.head()
```

```python
# Function to convert km/l to miles/gallon
def kmpl_to_mpg(kmpl):
    return kmpl * 0.621371 * 0.264172 if isinstance(kmpl, (int, float)) else None

# Add new column with the conversion
data['Mileage_in_miles'] = data['Fuel_Efficiency'].apply(kmpl_to_mpg)

# Display the updated data
data.head()
```

```python
# Arrange: Sort data by 'Car_Age' in ascending order and 'Car_Price' in descending order
arranged_data = data.sort_values(by=['Car_Age', 'Car_Price'], ascending=[True, False])

# Display the arranged data
arranged_data.head()
```

```python
# Group by 'Owner_Type' and calculate count, mean, and sum for 'Kilometers_Driven' and 'Seats'
summary_data = data.groupby('Owner_Type').agg({
    'Kilometers_Driven': ['count', 'mean', 'sum'],   # Count, mean, and sum of 'Kilometers_Driven'
    'Seats': ['mean', 'sum']                         # Mean and sum of 'Seats'
}).reset_index()

# Display the summarized data
print(summary_data.head())

```
